널값 처리 방법
1. 널값을 가진 데이터를 제거한다 (단.. 나중에 test 셋이나 production 단계에서 널값이 들어올 경우 문제 발생)
2. 널값을 특정값으로 채워넣는다
  - 가장 많은 값/평균 값으로 채워넣는다
  - 0값으로 채워넣는다 (주로 base값 혹은 없는 값을 의미하니깐...)
  - 별도의 값을 널값에 맵핑한다 (ex. -1)
3. one-hot encoding 등에 무엇에도 해당하지 않는다로 표시하여 널값임을 나타낸다.

```python
def count_characters(series):
    series = series.dropna()
    series = series.apply(lambda x: list(x))
    series = pd.DataFrame(list(series.values))
    
    vc = None
    for c in series.columns:
        if vc is None:
            vc = series[c].value_counts()
        else:
            vc += series[c].value_counts()
    return vc
```

```python
count_characters(train_df['nom_5']).sort_index()
```
